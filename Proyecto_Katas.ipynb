
1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados

```python
def frecuencia_letras(texto: str) -> dict:
    """
     Devuelve un diccionario con la frecuencia de cada letra en la cadena.
     No cuenta los espacios.
    """
    frecuencias = {}
    for caracter in texto:
        if caracter != " ":  # Ignorar espacios
            frecuencias[caracter] = frecuencias.get(caracter, 0) + 1
    return frecuencias
print(frecuencia_letras("hola mundo"))
```

4. Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función map()


```python
def diferencia_listas(lista1, lista2):
    """
    Calcula la diferencia entre las dos listas
    
    Parametros:
    lista1 (list): La primera lista de números
    lista2 (list): La segunda lista de números  
    Retorna:
    list: Una nueva lista con los elementos que están en lista1 pero no en lista2
    """
    return [x for x in lista1 if x not in lista2]
```

```python
notas = [6, 7, 8, 4, 5, 9]
nota_aprobado = 5

def calcular_notas(notas, nota_aprobado=5):
    promedio = sum(notas) / len(notas)
    estado = "aprobado" if promedio >= nota_aprobado else "suspenso"
    return (promedio, estado)

resultado = calcular_notas(notas, nota_aprobado)
print(f"Promedio: {resultado[0]}, Estado: {resultado[1]}")
```

6.Escribe una función que calcule el factorial de un número de manera recursiva.


```python
def factorial(n):
    """
    Calcula el factorial de un número de manera recursiva.

    Parámetros:
        n (int): Número entero no negativo.

    Retorna:
        int: El factorial de n (n!).
    """
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(5))
```

```python
class Arbol:

    def __init__(self):
        self.tronco = 1        # tronco inicia en longitud 1
        self.ramas = []        # lista vacía de ramas

    def crecer_tronco(self):
        self.tronco += 1

    def nueva_rama(self):
        self.ramas.append(1)   # cada nueva rama empieza con longitud 1

    def crecer_ramas(self):
        self.ramas = [r + 1 for r in self.ramas]

    def quitar_rama(self, indice):
        if 0 <= indice < len(self.ramas):
            return self.ramas.pop(indice)
        else:
            raise IndexError("Índice de rama inválido")

    def info_arbol(self):
        return {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitudes_ramas": self.ramas
        }

```

Código a seguir:
1. Inicializar un usuario con su nombre, saldo y si tiene o no cuenta corriente mediante True y False.
2. Implementar el método retirar_dinero para retirar dinero del saldo del usuario. Lanzará un error en caso de no
poder hacerse.
3. Implementar el método transferir_dinero para realizar una transferencia desde otro usuario al usuario actual.Lanzará un error en caso de no poder hacerse.
4. Implementar el método agregar_dinero para agregar dinero al saldo del usuario.

Caso de uso:
1. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.
2. Agregar 20 unidades de saldo de "Bob".
3. Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".
4. Retirar 50 unidades de saldo a "Alicia".


```python
class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta_corriente):
        if cuenta_corriente not in (True, False):
            raise ValueError("El atributo 'cuenta_corriente' debe ser True o False.")

        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser positiva.")
        if cantidad > self.saldo:
            raise ValueError(f"{self.nombre} no tiene suficiente saldo para retirar.")
        self.saldo -= cantidad
        return self.saldo

    def transferir_dinero(self, otro_usuario, cantidad):
        """
        Realiza una transferencia DESDE otro_usuario HACIA self.
        """
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser positiva.")
        if cantidad > otro_usuario.saldo:
            raise ValueError(f"{otro_usuario.nombre} no tiene suficiente saldo para transferir.")

        otro_usuario.saldo -= cantidad
        self.saldo += cantidad

        return self.saldo, otro_usuario.saldo

    def agregar_dinero(self, cantidad):
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser positiva.")
        self.saldo += cantidad
        return self.saldo
```

```python
# 1. Crear dos usuarios: "Alicia" con saldo inicial de 100 y "Bob" con saldo inicial de 50, ambos con cuenta corriente.

cuenta1 = UsuarioBanco("Alicia", 100, True)
cuenta2 = UsuarioBanco("Bob", 50, True) 
```

```python
# 2. Agregar 20 unidades de saldo de "Bob".

cuenta2.agregar_dinero(20)
```

```python
# 3. Hacer una transferencia de 80 unidades desde "Bob" a "Alicia".

cuenta2.transferir_dinero(cuenta1, 80)
```

```python
# 4. Retirar 50 de Alicia

cuenta1.retirar_dinero(50)
```

37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras ,
reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentrode la función procesar_texto.

Código a seguir:
1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene que devolver un diccionario.
2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva . Tieneque devolver el texto con el remplazo de palabras.
3. Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra eliminada.
4. Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un número de argumentos variable según la opción indicada.

Caso de uso:
Comprueba el funcionamiento completo de la función procesar_texto

```python
import string
import re

# 1. Contar palabras
def contar_palabras(texto):
    texto = texto.lower()
    for c in string.punctuation:
        texto = texto.replace(c, '')
    palabras = texto.split()
    conteo = {}
    for palabra in palabras:
        conteo[palabra] = conteo.get(palabra, 0) + 1
    return conteo


# 2. Reemplazar palabras
def reemplazar_palabras(mi_texto, palabra_original, palabra_nueva):
    palabras = mi_texto.split()
    resultado = [palabra_nueva if palabra.lower() == palabra_original.lower() else palabra 
                 for palabra in palabras]
    return ' '.join(resultado)


# 3. Eliminar palabra
def eliminar_palabra(texto, palabra_a_eliminar):
    pattern = re.compile(rf"\b{re.escape(palabra_a_eliminar)}\b", flags=re.IGNORECASE)
    nuevo_texto = pattern.sub("", texto)
    return re.sub(r"\s+", " ", nuevo_texto).strip()


# 4. Función procesadora general
def procesar_texto(mi_texto, opcion, *args):
    if opcion == "contar":
        return contar_palabras(mi_texto)

    elif opcion == "reemplazar":
        if len(args) != 2:
            raise ValueError("Para reemplazar necesitas: palabra_original y palabra_nueva")
        return reemplazar_palabras(mi_texto, args[0], args[1])

    elif opcion == "eliminar":
        if len(args) != 1:
            raise ValueError("Para eliminar necesitas una palabra.")
        return eliminar_palabra(mi_texto, args[0])

    else:
        raise ValueError("Opción no válida. Usa: contar, reemplazar, eliminar")
```

```python
mi_texto = "Hola mundo. Estoy usando python"
```

40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" "circulo" o "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).


```python
def calcular_area(figura, datos):
    if figura == "rectangulo":
        largo, ancho = datos
        return largo * ancho
    elif figura == "circulo":
        radio, = datos
        return 3.1416 * radio ** 2
    elif figura == "triangulo":
        base, altura = datos
        return 0.5 * base * altura
    else:
        raise ValueError("Figura no reconocida. Usa 'rectangulo', 'circulo' o 'triangulo'.")
```

El programa debe hacer lo siguiente:
 1. Solicita al usuario que ingrese el precio original de un artículo.
 2. Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).
 3. Si el usuario responde que sí, solicita que ingrese el valor del cupón de descuento.
 4. Aplica el descuento al precio original del artículo, siempre y cuando el valor del cupón sea válido (es decir, mayor a cero). Por ejemplo, descuento de 15€.
 5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin él.

```python
precio_articulo = input("Ingresa el precio del artículo: ")

try:
    precio_articulo = float(precio_articulo)

    tiene_cupon = input("¿Tienes un cupón de descuento? (sí/no): ").strip().lower()

    descuento = 0

    if tiene_cupon in ("si", "sí"):
        valor_cupon = float(input("Ingresa el valor del cupón: "))
        if valor_cupon > 0:
            descuento = valor_cupon

    precio_final = precio_articulo - descuento
    if precio_final < 0:
        precio_final = 0  # evitar precios negativos

    print(f"El precio final es: {precio_final} €")

except ValueError:
    print("Por favor, ingresa valores numéricos válidos.")

```
